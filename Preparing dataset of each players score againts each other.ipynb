```python
import pandas as pd
import numpy as np
import time
from datetime import datetime


ball=pd.read_csv('C:\\Users\\adity\\OneDrive\\Documents\\Python Scripts\\Research Paper\\IPL_Ball_by_Ball_2008_2021.csv')
match=pd.read_csv('C:\\Users\\adity\\OneDrive\\Documents\\Python Scripts\\Research Paper\\IPL_Matches_2008_2021.csv')

def all_teams():
    x=[]
    for i in match['Team1Players']:
        if i not in x:
            x.append(i)

    for i in match['Team2Players']:
        if i not in x:
            x.append(i)
    return x

def chemistry(name1,name2):
    x=all_teams()
    c=0
    if name1==name2:
        return 0
    for i in x:
        y=i.split("'")[1::2]
        #print(type(i))
        #print(type(y),y)
        if name1 in y and name2 in y:
            c=c+1
            #display(y)
    #print(c)
    return c

def all_players(x):
    y=set({})
    for i in x:
        zx=i.split("'")
        for j in zx:
            if j!=', ' and j!='[' and j !=']':
                y.add(j)
    return list(y)


def chem_all():
    x=all_players(all_teams())
    d=dict({'Player1':[],'Player2':[],'Chemistry':[]})
    for i in range(0,len(x)-1):
        for j in range(i+1,len(x)):
            d['Player1'].append(x[i])
            d['Player2'].append(x[j])
            d['Chemistry'].append(chemistry(x[i],x[j]))
            print(i,j,x[i],x[j])
            
    return pd.DataFrame(d)
#chemistry_all_players=chem_all()

def both_teams():
    x=[]
    y=[]
    for i in match['Team1Players']:
        x.append(i)

    for i in match['Team2Players']:
        y.append(i)
    return x,y

def against_chem(name1,name2):
    x,y=both_teams()
    c=0
    for i in range(len(x)):
        zx=x[i].split("'")[1::2]
        zy=y[i].split("'")[1::2]
        if (name1 in zx and name2 in zy) or (name1 in zy and name2 in zx):
            c=c+1
            
        #if :
            #c=c+1
            #display(zx,zy)
    #print(name1,name2,c)
    return c

def bat_bowl():
    bat=[]
    bowl=[]
    for i in ball['batter']:
        if i not in bat:
            bat.append(i)

    for i in ball['bowler']:
        if i not in bowl:
            bowl.append(i)
    #display(bat,bowl)
    return bat,bowl

def stats_compare(name1,name2,n):
    #print("Start",datetime.now())
    bat,bowl=bat_bowl()
    p1score=0
    p2score=0
    
    #n=against_chem(name1,name2)
    #print("END !",datetime.now())
    #print()
    
    if name1 in bat and name2 in bowl:
        df1=ball[ball['batter']==name1]
        df2=df1[df1['bowler']==name2]
        #print(len(df2))
        if(len(df2)==0):
            p1score=0
            p2score=0
        else:
            out=len(df2[df2['isWicketDelivery']==1])
            balls_played=len(df2)
            total_run = sum(df2['batsman_run'])
            try:
                p1score+=total_run/balls_played
            except:
                p1score+=total_run/1
            try:
                p2score+=out/n
            except:
                p2score+=out/1
    #print("END 1",datetime.now())
    #print(p1score,p2score)
    #print()
    if name1 in bowl and name2 in bat:
        df1=ball[ball['batter']==name2]
        df2=df1[df1['bowler']==name1]
        #print(len(df2))
        if(len(df2)==0):
            p1score=0
            p2score=0
        else:
            out=len(df2[df2['isWicketDelivery']==1])
            balls_played=len(df2)
            total_run = sum(df2['batsman_run'])
            try:
                p2score+=total_run/balls_played
            except:
                p2score+=total_run/1
            try:
                p1score+=out/n
            except:
                p1score+=out/1
    #print("End 2",datetime.now())
    #print(p1score,p2score)
    #print()
    #print()
    #print(name1,name2,p1score,p2score)
    return p1score,p2score


def players():
    x=all_teams()
    y=set({})
    for i in x:
        zx=i.split("'")[1::2]
        for j in zx:
            y.add(j)
    return list(y)

```
