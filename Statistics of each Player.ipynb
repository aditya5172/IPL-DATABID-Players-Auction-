# IPL Each Player's Statistics

## Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

## Importing Required Datasets

```python
df=pd.read_csv('C:\\Users\\adity\\OneDrive\\Documents\\Python Scripts\\Research Paper\\IPL_Ball_by_Ball_2008_2021.csv')
df2=pd.read_csv('C:\\Users\\adity\\OneDrive\\Documents\\Python Scripts\\Research Paper\\IPL_Matches_2008_2021.csv')
deliveries=df.copy()
dfd=df.copy()
```

```python
years=['2015','2016','2017','2018','2019','2020','2021']
k=0
for i in years:
    dd=df2.query('Season==@i')
    for j in dd['ID']:
        if k==0:
            k=k+1
            x=df.query('ID==@j')
            
        else:
            x=pd.concat([x,df.query('ID==@j')],axis=0)
```

### Batsmen Statistics Table

```python
g=0
ba={'Name':[''],'Runs':[0],'Fours':[0],'Six':[0],'Singles':[0],'Doubles':[0],'Average':[0.0],
    'Strike_rate':[0.0],'Balls_faced':[0],'Matches_played':[0],'man_of_the_match':0,'extra_points':[0],'Not Out':[0]}

bat=pd.DataFrame(ba)

def n(x):
    ba={'Name':x,'Runs':0,'Fours':0,'Six':0,'Singles':0,'Doubles':0,
        'Average':0.0,'Strike_rate':0.0,'Balls_faced':0,'Matches_played':0,'man_of_the_match':0,'extra_points':0,'Not Out':0}

    return ba

for i in dfd['batter'].unique():
    bat=bat.append(n(i),ignore_index=True)
bat
```

### Runs

```python
g=0

x2=pd.DataFrame(dfd.groupby(['batter'])['batsman_run'].sum())
for i in range(1,len(bat['Name'])):
    x=bat['Name'][i]
    y=x2.query('batter==@x')
    bat.loc[i,'Runs']=list(list(y.values)[0])[0]
        
bat
```

### Singles, Doubles, Fours and Sixes Count

```python
g=0

for i in range(1,len(bat['Name'])):
    x=bat['Name'][i]
    dd=dfd.query('batter==@x')
    for j in dd.index:
        if dd['batsman_run'][j]==4:
            y=bat['Fours'][i]+1
            bat.loc[i,'Fours']=y
        elif dd['batsman_run'][j]==6:
            z=bat['Six'][i]+1
            bat.loc[i,'Six']=z
        elif dd['batsman_run'][j]==1:
            a=bat['Singles'][i]+1
            bat.loc[i,'Singles']=a
        elif dd['batsman_run'][j]==2:
            b=bat['Doubles'][i]+1
            bat.loc[i,'Doubles']=b
            
        bat.loc[i,'Balls_faced']=bat['Balls_faced'][i]+1

bat
```

### Strike Rate

```python
g=0
n=0
y=[]
q=0
for i in range(1,len(bat['Name'])):
    x=bat['Name'][i]
    r=(int)(bat['Runs'][i])
    bow=(int)(bat['Balls_faced'][i])
    sr=(r/bow)*100
    bat.loc[i,'Strike_rate']=sr
bat
```

### Matches Played

```python
g=0
n=0
y=[]
q=0
zz=0

for i in range(0,len(dfd['ID'])):
    x=deliveries['batter'][i]
    m=len(bat)
    
    if n==deliveries['ID'][i]:
        for k in range(0,len(y)):
             if y[k]==x:
                zz=1

        if zz==0:
            y.append(x)
            for jj in range(0,m):
                if bat['Name'][jj]==x:
                    bat.loc[jj,'Matches_played']=bat['Matches_played'][jj]+1
        zz=0
            
    elif n!=deliveries['ID'][i]:
        y=[]
        y.append(deliveries['batter'][i])
        n=deliveries['ID'][i]
        for jj in range(0,m):
             if bat['Name'][jj]==x:
                bat.loc[jj,'Matches_played']=bat['Matches_played'][jj]+1

bat
```

### Not Outs

```python
for i in range(0,len(dfd['ID'])):
    x=deliveries['overs'][i]
    y=deliveries['ballnumber'][i]
    z=deliveries['isWicketDelivery'][i]
    b=deliveries['batter'][i]
    ns=deliveries['non-striker'][i]
    m=len(bat)
    if x==19 and y==6:
        if z==0:
            for jj in range(0,m):
                if bat['Name'][jj]==b:
                    xxx=bat['Not Out'][jj]+1
                    bat.loc[jj,'Not Out']=xxx
                if bat['Name'][jj]==ns:
                    xxx=bat['Not Out'][jj]+1
                    bat.loc[jj,'Not Out']=xxx
        else:
            for jj in range(0,m):
                if bat['Name'][jj]==b and deliveries['player_out'][i]!=b:
                    xxx=bat['Not Out'][jj]+1
                    bat.loc[jj,'Not Out']=xxx
                if bat['Name'][jj]==ns and deliveries['player_out'][i]!=ns:
                    xxx=bat['Not Out'][jj]+1
                    bat.loc[jj,'Not Out']=xxx
                    
bat
```

### Average

```python
for i in range(1,len(bat['Name'])):
    x=bat['Matches_played'][i]-bat['Not Out'][i]
    if x<1:
        x=1
    bat.loc[i,'Average']=(bat['Runs'][i]/x)
bat
```

## Bowlers Statistics Table

```python
dfd2=dfd
bo={'Name':[''],'Runs Conceded':[0],'Bowling Average':[0.0],'Strike_rate':[0.0],'Balls_Bowled':[0],'Wickets':[0],'Economy':[0],
    'Matches_played':[0],'man_of_the_match':[0],'extra_points':[0],'Overs_Bowled':[0.0],'No_balls_Bowled':[0]}

bowl=pd.DataFrame(bo)

def nn(x):
    bo={'Name':x,'Runs Conceded':0,'Bowling Average':0.0,'Strike_rate':0.0,'Balls_Bowled':0,'Wickets':0,'Economy':0,
    'Matches_played':0,'man_of_the_match':0,'extra_points':0,'Overs_Bowled':0.0,'No_balls_Bowled':0}

    return bo
for i in dfd2['bowler'].unique():
    bowl=bowl.append(nn(i),ignore_index=True)
bowl
```

### Runs Conceded

```python
x2=pd.DataFrame(dfd2.groupby(['bowler'])['batsman_run'].sum())
for i in range(1,len(bowl['Name'])):
    x=bowl['Name'][i]
    y=x2.query('bowler==@x')
    bowl.loc[i,'Runs Conceded']=list(list(y.values)[0])[0]
bowl
```

### Bowls Bowled

```python
x2=pd.DataFrame(dfd2.groupby('bowler')['bowler'].count())
for i in range(1,len(bowl['Name'])):
    x=bowl['Name'][i]
    y=x2[x2.index==x]
    bowl.loc[i,'Balls_Bowled']=list(list(y.values)[0])[0]
bowl
```

### Overs Bowled

```python
for i in range(1,len(bowl['Name'])):
    x=''
    x=str(int(bowl['Balls_Bowled'][i]/6)) + '.'+ str(bowl['Balls_Bowled'][i]%6)
    x=float(x)
    bowl.loc[i,'Overs_Bowled']=x
bowl
```

### Matches Played

```python
g=0
n=0
y=[]
q=0
zz=0

for i in range(0,len(dfd2['ID'])):
    x=dfd2['bowler'][i]
    m=len(bowl)

    if n==dfd2['ID'][i]:
        for k in range(0,len(y)):
             if y[k]==x:
                zz=1
        if zz==0:
            y.append(x)
            for jj in range(0,m):
                if bowl['Name'][jj]==x:
                    bowl.loc[jj,'Matches_played']=bowl['Matches_played'][jj]+1
        zz=0
    elif n!=dfd2['ID'][i]:
        y=[]
        y.append(dfd2['bowler'][i])
        n=dfd2['ID'][i]
        for jj in range(0,m):
             if bowl['Name'][jj]==x:
                bowl.loc[jj,'Matches_played']=bowl['Matches_played'][jj]+1
bowl
```

### Wickets Taken

```python
for i in range(0,len(dfd2['ID'])):
    x=dfd2['bowler'][i]
    y=dfd2['kind'][i]
    if y=='caught' or y == 'lbw' or y == 'bowled' or y == 'caught and bowled' or y=='stumped' or y=='hit wicket':
        for j in range(0,len(bowl['Name'])):
            if bowl['Name'][j]==x:
                bowl.loc[j,'Wickets']=bowl['Wickets'][j]+1
bowl
```

### Bowling Average, Strike rate and Economy

```python
for i in range(0,len(bowl['Name'])):
    x=bowl['Wickets'][i]
    if bowl['Wickets'][i]<1:
        x=1
    
    bowl.loc[i,'Bowling Average']=bowl['Runs Conceded'][i]/x

    bowl.loc[i,'Strike_rate']=bowl['Balls_Bowled'][i]/x
    bowl.loc[i,'Economy']=bowl['Runs Conceded'][i]/bowl['Overs_Bowled'][i]
bowl
```
